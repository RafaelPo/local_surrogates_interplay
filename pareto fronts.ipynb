```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns; sns.set()

import numpy as np
```

```python
mpl.rcParams['lines.linewidth'] = 3

fig, ax = plt.subplots(figsize=(10, 10))

x = np.arange(0, 3, 1e-3)
y = np.sqrt(1 - x**2)
ax.plot(x, y, label='Pareto Front for Global Surrogate')

c = 1.25
v = 1.05
q = 1

y = v*np.sqrt(q - (x/c)**2)
ax.plot(x, y, label='Pareto Front for Local Surrogate')

y = c*np.sqrt(q - (x/v)**2)
ax.plot(x, y, label='Pareto Front for Local Surrogate')

ax.set_xlabel('Accuracy', fontsize=24)
ax.set_ylabel('Simplicity\n(Inverse Complexity)', fontsize=24)


ax.arrow(0.375, 0.95, 0.065, 0.14, color='k', length_includes_head=True,
          head_width=0.03, head_length=0.02)

ax.arrow(0.95, 0.375, 0.14, 0.065, color='k', length_includes_head=True,
          head_width=0.03, head_length=0.02)

plt.legend(fontsize=16)
_ = plt.setp(ax.get_xticklabels(), visible=False)
_ = plt.setp(ax.get_yticklabels(), visible=False)

plt.tight_layout()
plt.savefig('pareto_fronts.png', dpi=200)
```
